# Ejercicio 2
Escriba una función de Python para dar la suma de los números de una lista.

````python
def suma(lista_numeros):
    """Esta funcion va a realizar la suma de los elementos de una lista"""
    suma = 0
    for numero in lista_numeros:
        suma = suma + numero
    return suma
````

````python
lista_numeros = [1111, 63123, 1423432, 4123123]
print(suma(lista_numeros))
````

# Ejercicio 3

Dada una lista y un número, cree una función que devuelva dos parámetros:

1) Si está dicho número en esa lista. En dicho caso Devolverá `TRUE`. En caso contrario, `False`.

2) Si la longitud de la lista coincide con dicho número. En dicho caso devolverá `TRUE`. En caso contrario, `False`.

````python
def doble_salida(lista_numeros, numero_comprobar):
    esta_num_presente = True if numero_comprobar in lista_numeros else False
    longitud_igual = True if len(lista_numeros) == numero_comprobar else False
    return esta_num_presente, longitud_igual

print(doble_salida([1,2,3,4,5], 5))
````

````python
def funcion(lista,numero):
    numero_presente = numero in lista
    longitud = len(lista) == numero
    return numero_presente, longitud

funcion([1,2,3,4,5], 6)
````

# Ejercicio 4

Dada una lista, cree dos funciones, una que devuelva el mayor número y otra que devuelva en menor número.



````python
def retornar_mayormenor(lista):
    mayor=lista[0]
    menor=lista[0]
    for x in range(1,len(lista)):
        if lista[x]>mayor:
            mayor=lista[x]
        else:
            if lista[x]<menor:
                menor=lista[x]
    return [mayor,menor]

lista=[1,2,3,4,5]
rango=retornar_mayormenor(lista)
print("Mayor numero de la lista:",rango[0])
print("Menor numero de la lista:",rango[1])
````

# Ejercicio 8

Cree una función para contar la cantidad de números impares de una lista.


````python
lista_numeros3 = [4, 7, 19, 34, 69, 102, 105]

def contar_impares(lista_numeros3):
    contador = 0
    for numero in lista:
        if numero % 2 != 0:
            contador +=2

    return contador

cantidad_impares = contar_impares(lista_numeros3)

print("Cantidad de números impares en la lista:", cantidad_impares)
````

````python
def contar_impares_usando_count(lista):
    lista_booleanos = [num % 2 != 0 for num in lista]
    return lista_booleanos.count(True)

lista_ejemplo = [4, 7, 19, 34, 69, 102, 105]
contar_impares_usando_count(lista_ejemplo)

````
